```python
from sentence_transformers import SentenceTransformer
encoder=SentenceTransformer("all-mpnet-base-v2")
vector=encoder.encode(df['text'])
print("Vector",vector)
print("vector shape",vector.shape)
```

```python
dim=vector.shape[1]
dim
```

```python

import faiss
index=faiss.IndexFlatL2(dim)
index
```

```python
index.add(vector)
```

```python
search_query="I want to go to pakistan"
vec=encoder.encode(search_query)
vec.shape
```

```python
import numpy as np
svec=np.array(vec).reshape(1,-1)
svec.shape
```

```python
distance,I=index.search(svec,k=2)   # k means how many rows similar to search_query you want to return
print(distance)
print(I)

```
